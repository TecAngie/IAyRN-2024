```python
"""
Nombre: Angie Sofia Hernández Tristán.
Clase: Inteligencia Artificial, LMV, N1.
Título: Actividad 3 "Solución de problemas".
"""
# 1. Cargar librerias
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# 2. Cargar datos
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100,1) * 2
```

```python
# 3. Preprocesamiento de datos
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=42)
# El parámetro 0.2 indica el 20% de los datos que se usaran para la prueba, el
# restante (80%) se usará para entrenamiento %% 4.Selección y entrenamiento
# del modelo
```

```python
# 4.Selección y entrenamiento del modelo
model = LinearRegression()
model.fit(train_data, train_labels)
```

```python
# 5. Prueba de modelo
predictions = model.predict(test_data)
mse = mean_squared_error(test_labels, predictions)
```

```python
plt.scatter(test_data, test_labels, color='black')
plt.plot(test_data, predictions, color='blue', linewidth=3)
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.title('Modelo de regresión lineal')
plt.show()

print("Error cuadrático medio:", mse)
```
